```python
# Importing the packages
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# initialising the CNN,Convolution,Pooling
classifier = Sequential()

# Step 1 - Convolution
classifier.add(Conv2D(32, (3, 3), input_shape = (173, 166, 3), activation = 'relu'))
# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
# Adding a second convolutional layer
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

```

```python
# Step 3 - Flattening
classifier.add(Flatten())
```

```python
# Step 4 - Full connection

classifier.add(Dense(units = 512, activation = 'relu'))
classifier.add(Dense(units = 256, activation = 'relu'))
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 32, activation = 'relu'))
classifier.add(Dense(units = 16, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# step5
# Compiling the CNN
classifier.compile(optimizer = 'adam',loss = 'binary_crossentropy', metrics = ['accuracy'])
# adam = adaptive learning rate optimization algorithm
#  it computes individual learning rates for different parameters
```

```python
# Part 2 - Fitting the CNN to the images
# This is also called Image Augmentation
# Preprocessing the image to prevent overfitting(good result in the training set but 
# poor result in test set)

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                    shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
batch_size = 16

training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/copiedmangoset/Trainingdata',
                                                 target_size = (173, 166),
                                                 batch_size = 16,
                                                 class_mode = 'binary')
```

```python
# bach size =  number of training examples utilized in one iteration
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/copiedmangoset/Testdata',
                                            target_size = (173, 166),
                                            batch_size = 16,
                                            class_mode = 'binary')
```

```python
classifier.fit_generator(training_set,
                         steps_per_epoch =878/batch_size, #no of training image
                         epochs = 10,
                         validation_data = test_set,
                         validation_steps = 519/batch_size) #no of test image
classifier.save("/content/drive/MyDrive/XMVO.h5")
```

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('/content/drive/MyDrive/copiedmangoset/Testdata/GoodMangoes/goodmangoestest (101).jpg',
                                     target_size = (173,166))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image,axis = 0)
result = classifier.predict(test_image)


if result[0][0]==1:
    prediction = 'GoodMangoes'
    print(prediction)
else:
    prediction='Badmangoes'
    print(prediction)
    
```
